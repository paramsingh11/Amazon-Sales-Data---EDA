```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("amazon.csv")
df.head(5)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
summary_stats = df.describe()
print(summary_stats)
```

```python
def convert_to_numeric(column):
    return pd.to_numeric(column.replace('[₹,%,]', '', regex=True))
```

```python
df['discounted_price'] = convert_to_numeric(df['discounted_price'])
df['actual_price'] = convert_to_numeric(df['actual_price'])
df['discount_percentage'] = convert_to_numeric(df['discount_percentage'])
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = convert_to_numeric(df['rating_count'])
```

```python
df['rating_count'].fillna(df['rating_count'].median(), inplace=True)
```

```python
df_cleaned = df.drop_duplicates()
```

```python
df_cleaned.info()
```

```python
df_cleaned.head()
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['rating'], kde=True, bins=10)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['discounted_price'], kde=True, bins=20)
plt.title('Distribution of Discounted Prices')
plt.xlabel('Discounted Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['actual_price'], kde=True, bins=20)
plt.title('Distribution of Actual Prices')
plt.xlabel('Actual Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='discounted_price', y='category', data=df_cleaned)
plt.title('Discounted Price by Category')
plt.xlabel('Discounted Price')
plt.ylabel('Category')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['discount_percentage'], kde=True, bins=20)
plt.title('Distribution of Discount Percentages')
plt.xlabel('Discount Percentage')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
category_counts = df_cleaned['category'].value_counts()
sns.barplot(y=category_counts.index, x=category_counts.values)
plt.title('Number of Products in Each Category')
plt.xlabel('Number of Products')
plt.ylabel('Category')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
top_reviewers = df_cleaned['user_name'].value_counts().head(10)
sns.barplot(y=top_reviewers.index, x=top_reviewers.values)
plt.title('Top 10 Reviewers by Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('User Name')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
corr_matrix = df_cleaned[['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
